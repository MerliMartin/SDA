### Sort the data frame by the number of points obtained for the exam, from highest to lowest. In case of the same number of points, the persons should be displayed in alphabetical order.

```python
student_scores_data = {
    "Names": ['Merli', 'Aigi', 'Olga', 'Tartin', 'Indrek', 'Andres', 'Kerly', 'Merili', 'Sten', 'Niko'],
    "Scores": [100, 75, 85, 45, 50, 55, 73, 89, 45, 100]
}
student_score_table = pd.DataFrame(student_scores_data)  # Created dataframe about students and their exams scores
student_score_table
```

```python
sorted_table = student_score_table.sort_values(['Scores', 'Names'], ascending=[False, True])
sorted_table
```

```python
sorted_table.reset_index(drop=True)
```
