### Add a third column to the box which will contain the percentage of student scores on this exam (that is, each numerical value must be divided by the maximum score that could be obtained on this exam). Next, we want to anonymize the first name column: we want only the first and last letters of the first name to be left. For example, for the name "John" we want to keep "J ... n"


```python
student_scores_data = {
    "Names": ['Merli', 'Aigi', 'Olga', 'Martin', 'Indrek', 'Andres', 'Kerly', 'Merili', 'Sten', 'Niko'],
    "Scores": [100, 75, 85, 45, 50, 55, 73, 89, 94, 100]
}
student_score_table = pd.DataFrame(student_scores_data)  # Created dataframe about students and their exams scores
student_score_table
```

```python
max_score = 100
```

```python
student_score_table["%_of_scores"] = student_score_table['Scores'].apply(lambda score_num: score_num/max_score)
student_score_table
```

```python
student_score_table['Names'] = student_score_table.Names.apply(lambda std_name: (std_name[0]+ '...' + std_name[-1]))
student_score_table
```
