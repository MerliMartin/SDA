**Task 1
Load the file ves-usd.csv from the ML-datasets repository: https://github.com/matzim95/ML-datasets and plot a linear graph to show the dependence between bolivar and US dollar.**

```python
# displaying plots in compilation
%matplotlib inline  
import matplotlib.pyplot as plt  # convention

import pandas as pd
```

```python
ves_usd = pd.read_csv('../datas/ves-usd.csv', thousands=',', index_col='Date', parse_dates=True)
ves_usd.head()
```

```python
fig, ax = plt.subplots()

ax.plot(ves_usd.index, ves_usd['Rate'])

plt.show()
```
