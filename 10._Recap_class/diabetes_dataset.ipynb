```python
# Load dataset
from sklearn.datasets import load_diabetes  # Regression

# Import libraries 
import numpy as np
import pandas as pd
```

```python
data = load_diabetes(as_frame=True)
df = data.data
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.sex.unique()
```

```python
df["sex_cat"] = df.sex.apply(lambda x: "male" if x > 0 else "female")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].boxplot(df[df.sex_cat == "male"].bmi)
axs[1].boxplot(df[df.sex_cat == "female"].bmi)
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
sns.scatterplot(
    df,
    x="bmi",
    y="bp",
    hue="sex"
).set(title="Relationship between BMI and BP")
```

```python
sns.regplot(
    df,
    x="bmi",
    y="bp"
).set(title="Relationship between BMI and BP")
```
