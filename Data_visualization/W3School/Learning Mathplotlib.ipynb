### Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting

```python
xpoints1 = np.array([1,8])
ypoints1 = np.array([3,10])
```

```python
plt.plot(xpoints1,ypoints1)
plt.show()
```

**Plotting without a line**<br>
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
plt.plot(xpoints1,ypoints1, 'o')
plt.show()
```

**Multiple points**

```python
xpoints_longer = np.array([1,2,6,8])
ypoints_longer = np.array([3,8,1,10])
```

```python
plt.plot(xpoints_longer,ypoints_longer)
plt.show()
```

### Markers

```python
plt.plot(ypoints_longer, marker='*') # marker = 'o', '|', '*' - Marker Reference
plt.show()
```

**Format strings fmt** <br>
*marker|line|color* <br>
*:r

```python
plt.plot(ypoints_longer, '*-.m')  # Line reference
plt.show()
```

**Marker size & Marker color (edge)**

```python
plt.plot(ypoints_longer, marker='*', ms=15, mec='r')  # marker size == ms, mec == maerkeredgecolor
plt.show()
```

```python
plt.plot(ypoints_longer, marker='*', ms=15, mfc='r')  # marker size == ms, mfc == markerfacecolor
plt.show()
```
