### Getting started with Matplotlib

```python
# displaying plots in compilation
%matplotlib inline  
import matplotlib.pyplot as plt  # convention

import pandas as pd  # library for data analysis

from pandas.plotting import register_matplotlib_converters

# using pandas units in matplotlib (e.g timeseries)
register_matplotlib_converters()
```

### First drawing

```python
# create a drawing (container for plots) and a plot (contains drawn data)
fig, ax = plt.subplots()


x = [1, 2, 3, 4]
y = [0, 1, 2, 3]
ax.plot(x, y)

ax.grid() # puts grid on chart backround

plt.show()
```

### Markers and lines in the chart

```python
n = 20
# creating a drawing
fig, ax = plt.subplots()
# drawing a plot using "axis" object method
# using markers - to emphasize the discreet nature of the data
ax.plot(weather.index[:n], weather['temp_max'][:n], marker='o', linestyle='--', color='r')
ax.plot(weather.index[:n], weather['temp_min'][:n], marker='v', linestyle='None', color='b')

plt.show()
```

### Plot parameters

```python
n = 31
# creating a drawing
fig, ax = plt.subplots()
# drawing a plot using "axis" object method
# using markers - to emphasize the discreet nature of the data
ax.plot(weather.index[:n], weather['temp_max'][:n], marker='o', linestyle='--', color='r')
ax.plot(weather.index[:n], weather['temp_min'][:n], marker='v', linestyle='dotted', color='b')

# X axis label
ax.set_xlabel('Date (year/month/day)')
# Y axis label
ax.set_ylabel('Temperature [centigrades]')
# plot title
ax.set_title('Temparature in Seattle - January 2012')
# data rotation on the abscissa axis
ax.xaxis.set_tick_params(rotation=60)

plt.show()
```

### Changing Y axis

```python
temps = pd.read_csv('../datas/temperature.csv', index_col='dt', parse_dates=True)
temps.index.names = ['Date']
temps
```

```python
merged = pd.merge(co2, temps, how='inner', on='Date')
merged
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('Time')

ax.plot(merged.index, merged['Interpolated'], color='b')
ax.set_ylabel('CO2', color='b')
ax.tick_params('y', colors='b')

# cloning the axix (mutal X axis, seperated Y)
ax2 = ax.twinx()
ax2.plot(merged.index, merged['AverageTemperature'], color='r')
ax2.set_ylabel('Temperature', color='r')
ax2.tick_params('y', colors='r')

plt.show()
```

### Bar charts

```python
population = pd.read_csv('../datas/population.csv', index_col='City')
population
```

```python
groupby_data = population.groupby('Sex')
both, female, male = [groupby_data.get_group(x) for x in groupby_data.groups]
both.head()
```

```python
fig, ax = plt.subplots()

ax.bar(both.index, both['Value'])
ax.xaxis.set_tick_params(rotation=90)

ax.set_ylabel('Population')
plt.show()
```

```python
# stacked barchart
fig, ax = plt.subplots()
ax.bar(male.index, male['Value'], label='Male')
ax.bar(both.index, both.Value, label='Both sexes', bottom=male.Value)

ax.tick_params('x', labelrotation=90)
ax.legend()

plt.show()
```

```python
medals = pd.read_csv('../datas/olympic.csv', index_col='ID')
medals
```

```python
def gropby_column(dataframe, column):
    groupby = dataframe.groupby(column)
    return {x: groupby.get_group(x) for x in groupby.groups}

all_sports = gropby_column(medals, 'Sport')
all_sports
```
