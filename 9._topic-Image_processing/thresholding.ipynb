```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read image

img = cv.imread("images/lenna.png")
```

```python
def bgr2rgb(img_array):
    img_rgb = img_array.copy()
    img_rgb[:, :, 0] = img[:, :,2]
    img_rgb[:, :, 2] = img[:, :, 0]
    return img_rgb
```

```python
def apply_filter(source, kernel):
    new_image = cv.filter2D(source, -1, kernel)
    print("New image shape: ", new_image.shape)
    if len(source.shape) == 3:
        return bgr2rgb(new_image)
    else:
        return new_image

def compare_original_img(img_original, img_filtered, label):
    fig, (ax1, ax2) = plt.subplots(1,2)
    
    ax1.imshow(img_original)
    ax1.set_title("Original Image")
    ax1.axis("off")
    
    ax2.imshow(img_filtered)
    ax2.set_title(label)
    ax2.axis("off")
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
x = 50
something, thresh1 = cv.threshold(gray_img, x, 255, cv.THRESH_BINARY)
compare_original_img(bgr2rgb(img), thresh1, label="Threshold Image")
```
