```python
import cv2 as cv
```

```python
# Read image

img = cv.imread("lenna.png")
img
```

```python
img.shape
```

```python
# Shrink image by 60%

scale_precent = 60  # percent of original size
width = int(img.shape[1] * scale_precent / 100)
height = int(img.shape[0] * scale_precent /100)
dim = (width, height)
dim
```

```python
# resize image

resized = cv.resize(img, dim, interpolation=cv.INTER_AREA)
```

```python
img.max(), resized.max()
```

```python
img.mean(), resized.mean()
```

```python
img.min(), resized.min()
```

```python
import matplotlib.pyplot as plt

# making a copy of image, to get right colors.
img_rgb = img.copy()

# reversing colors
img_rgb[:, :, 0] = img[:, :,2]
img_rgb[:, :, 2] = img[:, :, 0]

# making subplots to get images in the same plot
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(img)
ax1.axis("off")

ax2.imshow(img_rgb)
ax2.axis("off")
```

```python
plt.imshow(resized)
plt.axis("off")
```

```python
resized2 = cv.resize(img_rgb, dim, interpolation=cv.INTER_AREA)
```

```python
plt.imshow(resized2)
plt.axis("off")
```

```python
img_scaled = img_rgb/255.0

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(img_scaled)
ax1.set_title("Scaled Image")
ax1.axis("off")

ax2.imshow(img_rgb)
ax2.set_title("Original Image")
ax2.axis("off")
```

```python
print("{} to {}".format(img_scaled.min(), img_scaled.max()))
print("{} to {}".format(img_rgb.min(), img_rgb.max()))
```
