```python
import numpy as np
np.random.seed(42)
```

```python
def neuron(f):#here we defined it for 
    out = np.exp(-f)
    sol = np.round(1/(1+out),2) #hense the activation function is sigmoidal function
    return sol
```

<img src="Q1.png" alt="Drawing" style="width: 400px;"/>

```python
Input = np.array([[1],[-1]])
#Setting network parameters.
weight1 = np.array([[1,-1],[-2,1]])
bias1 = np.array([[1],[0]])
weight2 = np.array([[2,-2],[-1,-1]])
bias2 = np.array([[0],[0]])
weight3 = np.array([[3,-1],[-1,4]])
bias3 = np.array([[-2],[2]])

weights = [weight1,weight2,weight3]
bias = [bias1,bias2,bias3]
for i in range(len(weights)):
    #C = weights[i].dot(Input) + bias[i]
    C = np.dot(weights[i].transpose(),Input) + bias[i] #Here we are using matrix multipication as the functionality 
                                                       #In this feed forward computation of each layer.
    Input = neuron(C)
    print("Result after ",i+1,"st layer computation")
    print(Input)
Finaly = Input
print("Hence the computation of all the Hiden Layers is done and the result at the end of compuation is: ")
print(Finaly)
```

```python
C = int(input("Enter the size of input data you want to give: \n"))
D = np.array([[None]]*C)
for i in range(C):
    D[i] = int(input("number "+ str(i+1)+" : \n"))
D = np.int32(D)
print("Input data:\n",D)

A = int(input("How many number of layers you want to have in your architecture?\n"))
B = []
for i in range(A):
    s = int(input("Enter number of neurons you required in the layer_"+ str(i+1) +" :\n"))
    B.append(s)
B = np.int32(B)
print(B)
```
